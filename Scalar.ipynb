lakukan tahapan pada soal 1 dengan menggunakan tensorflow

```python
import tensorflow as tf
```

```python
#buat variabel skalar x dan simpan suatu nilai skalar
x_tf = tf.Variable(61, dtype=tf.int16) # dtype is optional
x_tf
```

```python
#tampilkan tipe data dari x
x_tf.numpy()
type(x_tf.numpy())
```

```python
x_tf.shape
```

```python
#buat variabel skalar y dan simpan suatu nilai skalar
y_tf = tf.Variable(10, dtype=tf.int16)
y_tf
```

```python
# tampilkan tipe data dari y
y_tf.numpy()
type(y_tf.numpy())
```

```python
y_tf.shape
```

```python
#jumlahkan x+y
x_tf + y_tf
```

```python
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
tf_sum.numpy()
```

```python
type(tf_sum.numpy())
```

```python
#buat variabel bernama x_float dan simpan suatu nilai desimal
x_tf_float = tf.Variable(0.25, dtype=tf.float16)
x_tf_float
```

```python
x_tf_float.shape
```
